```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, f1_score, roc_auc_score
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from imblearn.over_sampling import SMOTE
```

### âš–ï¸ Why We Use SMOTE

Real-world datasets are often **imbalanced**, meaning one class (e.g., "negative") has many more samples than the other class (e.g., "positive").  
This imbalance can cause machine learning models to become **biased towards the majority class**, leading to poor performance in detecting the minority class.

**SMOTE (Synthetic Minority Oversampling Technique)** helps solve this by:
- Generating **synthetic (artificial) samples** of the minority class, instead of just duplicating existing ones.  
- Creating new data points by interpolating between existing minority class samples.  
- Balancing the dataset, which allows the model to **learn better decision boundaries**.  

ðŸ‘‰ In short, SMOTE ensures that our classifier is not biased and can detect both classes more effectively.


```python
# 1. Apply SMOTE on full dataset

smote = SMOTE(random_state=42)
X_res, y_res = smote.fit_resample(X, y)

print("Before SMOTE:", y.value_counts().to_dict())
print("After SMOTE :", y_res.value_counts().to_dict())
```

```python
# 2. Train-Test Split

X_train, X_test, y_train, y_test = train_test_split(
    X_res, y_res, test_size=0.2, random_state=42, stratify=y_res
)
```

```python
# 3. Scaling for Logistic Regression

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

### Why We Scale Features for Logistic Regression

- **Logistic Regression** (and many other linear models) are sensitive to the **scale of input features**.  
- If features have very different ranges (e.g., `age` in years vs. `income` in thousands), the model may give **more weight** to features with larger values, even if they are not more important.  
- **StandardScaler** solves this by:
  - **Centering** the data â†’ subtracts the mean so each feature has mean = 0.  
  - **Scaling** the data â†’ divides by standard deviation so each feature has variance = 1.  

This ensures that:
- All features contribute **fairly** to the model.  
- The optimization algorithm used in Logistic Regression converges **faster and more reliably**.  

Note: Scaling is not always required for tree-based models (like Random Forest or XGBoost), but it is **very important** for Logistic Regression and other distance-based algorithms (e.g., SVM, KNN).


```python
# 4. Initialize Models

models = {
    "Logistic Regression": LogisticRegression(max_iter=2000, random_state=42),
    "Random Forest": RandomForestClassifier(n_estimators=300, max_depth=12, random_state=42),
    "XGBoost": XGBClassifier(n_estimators=300, learning_rate=0.1, max_depth=10, random_state=42, use_label_encoder=False, eval_metric="logloss")
}
```

```python
# 5. Train & Evaluate

results = {}

for name, model in models.items():
    if name == "Logistic Regression":
        model.fit(X_train_scaled, y_train)
        y_pred = model.predict(X_test_scaled)
        y_prob = model.predict_proba(X_test_scaled)[:, 1]
    else:
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)
        y_prob = model.predict_proba(X_test)[:, 1]

    acc = accuracy_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    roc = roc_auc_score(y_test, y_prob)

    results[name] = {"Accuracy": acc, "F1": f1, "ROC-AUC": roc}
```

```python
# 6. Print Results

for name, metrics in results.items():
    print(f"\n{name}:")
    for metric, value in metrics.items():
        print(f"  {metric}: {value:.4f}")
```

###  Model Interpretation with SHAP

To understand **why our XGBoost model makes certain predictions**, we use **SHAP (SHapley Additive exPlanations)**, a popular explainable AI method based on game theory.

- **`shap.TreeExplainer`**: Creates an explainer specifically designed for tree-based models like XGBoost, Random Forest, and LightGBM.  
- **SHAP values**: Show how much each feature **contributes (positively or negatively)** to a prediction compared to the average.  

#### Plots Generated:
1. **SHAP Summary Plot**  
   - Displays the overall importance of each feature across the dataset.  
   - Each point represents a SHAP value for one feature and one instance.  
   - The color indicates the feature value (low = blue, high = red).  
   - Helps identify the **most influential features** globally.  

2. **SHAP Waterfall Plot (for a single prediction)**  
   - Breaks down the contribution of each feature for **one prediction**.  
   - Starts from the base value (average model output) and adds/subtracts feature effects step by step.  
   - Helps understand **why the model predicted a certain class** for a specific instance.  

In short, SHAP helps us move beyond "black-box" predictions by providing both **global** (summary plot) and **local** (waterfall plot) explanations for XGBoost.


```python
import shap
import matplotlib.pyplot as plt

xgb_model = models['XGBoost']

print("\nXGBoost Feature Importance using SHAP:")

# Create TreeExplainer
explainer_shap = shap.TreeExplainer(xgb_model, X_train)

# Get SHAP values as Explanation object
shap_values = explainer_shap(X_test)

# Summary plot
shap.summary_plot(shap_values, X_test, feature_names=X_test.columns)

# Waterfall plot for a single prediction
shap.plots.waterfall(shap_values[0])
```

###  LIME for Model Interpretation

We also use **LIME (Local Interpretable Model-Agnostic Explanations)** to explain our modelâ€™s predictions.

- **`LimeTabularExplainer`**: A class from the `lime` library designed to work with **tabular datasets** (rows = samples, columns = features).  
- **How it works**:
  1. For a given instance (row), LIME perturbs the data by creating slightly modified samples.  
  2. It observes how the modelâ€™s predictions change with these perturbations.  
  3. Then it trains a simple interpretable model (like linear regression) on the perturbed samples.  
  4. The coefficients of this simple model show the **importance of features** for that single prediction.  

Unlike SHAP, which explains the model globally and locally, LIME focuses mainly on **local explanations** â€” why the model made a decision for a specific data point.


```python
import numpy as np

print("\nXGBoost Feature Contribution using LIME:")

explainer_lime = LimeTabularExplainer(
    training_data=np.array(X_train),
    feature_names=X_train.columns,
    class_names=['Class 0', 'Class 1'],
    mode='classification'
)

# Explain a single prediction
i = 0  # index of sample to explain
exp = explainer_lime.explain_instance(X_test.iloc[i], xgb_model.predict_proba, num_features=10)
exp.show_in_notebook(show_table=True)
```
